```python
# 7. Visualizar resultados
import pandas as pd
import glob

# Encontrar última pasta de resultados
result_dirs = sorted(glob.glob('comparison_results/RBF_Abrupt_Severe*'))
if result_dirs:
    latest = result_dirs[-1]
    print(f"\n=== RESULTADOS: {latest} ===\n")
    
    # Summary
    summary_file = f'{latest}/summary.txt'
    if os.path.exists(summary_file):
        with open(summary_file, 'r', encoding='utf-8') as f:
            print(f.read())
    
    # Tabela comparativa
    csv_file = f'{latest}/comparison_table.csv'
    if os.path.exists(csv_file):
        df = pd.read_csv(csv_file)
        print("\n=== TABELA DETALHADA ===\n")
        print(df[['chunk', 'model', 'accuracy', 'gmean']].to_string(index=False))
else:
    print("Nenhum resultado encontrado")
```
